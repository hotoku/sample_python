# データの生成と学習

```python
from sklearn.svm import SVC

import seaborn as sns
import sklearn.datasets
import sklearn.metrics
```

```python
blobs, y = sklearn.datasets.make_blobs(n_samples=200, centers=[(0.5,2.25), (1.0, 2.0)], cluster_std=0.3)
```

```python
sns.scatterplot(blobs[:,0], blobs[:,1], hue=y)
```

```python
clf = sklearn.svm.SVC(gamma="scale", probability=True)
fit = clf.fit(blobs, y)
```

# スコアを計算

```python
prob = fit.predict_proba(blobs)
```

# x, y座標を計算

```python
fpr, tpr, threshold = sklearn.metrics.roc_curve(y, prob[:,1])
```

```python
sns.lineplot(fpr, tpr)
```
